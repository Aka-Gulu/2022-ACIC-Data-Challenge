```python
import pandas as pd
import numpy as np

# Assuming you load your data into DataFrame
# Replace 'path_to_predictions.csv' and 'path_to_truth.csv' with your actual file paths
df_predictions = pd.read_csv('final_results.csv')
df_truth = pd.read_csv('ACIC_estimand_truths.csv')

# Filter necessary rows
filtered_predictions = df_predictions[(df_predictions['level'].isna()) & (df_predictions['year'].isna())]
filtered_truth = df_truth[(df_truth['variable'] == 'Overall') & (df_truth['level'].isna()) & (df_truth['year'].isna())]

filtered_predictions
```

```python
# Merge on 'dataset_num' or 'new_index' from predictions matching with 'dataset_num' in truth
merged_data = pd.merge(filtered_predictions, filtered_truth, left_on='new_index', right_on='dataset.num', how='inner')
print(merged_data.shape)
# Calculate RMSE
rmse = np.sqrt(np.mean((merged_data['satt'] - merged_data['SATT'])**2))
print("The RMSE is:", rmse)
```
